```python

#  2.4  Flow Control — if/elif/else, loops, pass/continue/break


sample_ratings = [8.8, 3.2, 6.5, 9.1, 4.8, 7.4, 5.0, 2.9, 7.8, 6.0]

# if / elif / else
def classify_movie(rating):
    """Classify a movie based on its TMDB rating."""
    if rating >= 8.0:
        return ' Masterpiece'
    elif rating >= 7.0:
        return ' Great'
    elif rating >= 5.5:
        return ' Average'
    elif rating >= 4.0:
        return ' Below Average'
    else:
        return ' Poor'

print("── Rating Classification ──")
for r in sample_ratings:
    print(f"  Rating {r:.1f} → {classify_movie(r)}")

# For loop with range
print("\n── For Loop (stars) ──")
for i in range(1, 6):
    print(f"  {'★' * i} ({i}/5)")

# While loop
print("\n── While Loop ──")
count = 0
high_rated = []
while count < len(sample_ratings):
    if sample_ratings[count] >= 7.0:
        high_rated.append(sample_ratings[count])
    count += 1
print(f"  High-rated movies (≥7.0): {high_rated}")

# Pass, Continue, Break
print("\n── Pass / Continue / Break ──")
for r in sample_ratings:
    if r < 3.0:   break      # stop at terrible movies
    if r < 5.0:   continue   # skip below-average
    pass                     # placeholder (no operation)
    print(f"  Processing rating: {r}")
```

```python

#  2.5  List & Dictionary Comprehensions


ratings = [8.8, 3.2, 6.5, 9.1, 4.8, 7.4, 5.0, 2.9, 7.8, 6.0]

# List comprehension
high_ratings   = [r for r in ratings if r >= 7.0]
scaled_ratings = [round(r * 10, 1) for r in ratings]   # scale to 100
labels         = [classify_movie(r) for r in ratings]

print("── List Comprehensions ──")
print(f"  Original ratings : {ratings}")
print(f"  High (≥7.0)      : {high_ratings}")
print(f"  Scaled (/100)    : {scaled_ratings}")
print(f"  Labels           : {labels}")

# Nested comprehension — rating matrix
matrix = [[round(i * j * 0.5, 1) for j in range(1, 4)] for i in range(1, 4)]
print(f"\n  3×3 Matrix       : {matrix}")

# Dictionary comprehension
movies_list = ['Inception','Interstellar','The Dark Knight','Avatar','Titanic']
r_list      = [8.8, 8.6, 9.0, 7.9, 7.8]

movie_rating_dict = {m: r for m, r in zip(movies_list, r_list)}
top_movies_dict   = {m: r for m, r in movie_rating_dict.items() if r >= 8.5}

print("\n── Dictionary Comprehensions ──")
print(f"  All    : {movie_rating_dict}")
print(f"  Top    : {top_movies_dict}")

# Set comprehension
genre_string = "action comedy drama action thriller drama sci-fi comedy"
unique_genres = {g.capitalize() for g in genre_string.split()}
print(f"\n── Set Comprehension ──")
print(f"  Unique genres: {unique_genres}")
```

```python

#  3.1  Defining Functions, Return, Scope


# Basic function
def get_movie_grade(rating: float) -> str:
    """Return letter grade for a TMDB rating."""
    if   rating >= 8.5: return 'A+'
    elif rating >= 7.5: return 'A'
    elif rating >= 6.5: return 'B'
    elif rating >= 5.0: return 'C'
    else:               return 'D'

# Function with default argument
def format_revenue(amount: float, currency: str = 'USD') -> str:
    """Format revenue as human-readable string."""
    if amount >= 1e9:
        return f"{currency} {amount/1e9:.2f}B"
    elif amount >= 1e6:
        return f"{currency} {amount/1e6:.1f}M"
    else:
        return f"{currency} {amount:,.0f}"

# Scope demo
GLOBAL_THRESHOLD = 7.0   # global variable

def is_recommended(rating):
    local_note = "Based on TMDB ratings"  # local variable
    return rating >= GLOBAL_THRESHOLD, local_note

# Test
test_ratings = [9.0, 7.5, 5.5, 3.2]
print("── Function Tests ──")
for r in test_ratings:
    rec, note = is_recommended(r)
    print(f"  Rating {r}: Grade={get_movie_grade(r)}  Recommended={rec}  [{note}]")

print(f"\n  Revenue 836836967  : {format_revenue(836836967)}")
print(f"  Revenue 2000000000 : {format_revenue(2_000_000_000, 'INR')}")
```

```python

#  3.2  *args and **kwargs


# *args — variable positional arguments
def movie_summary(*titles):
    """Print summary for any number of movies."""
    print(f"  Processing {len(titles)} movies:")
    for i, t in enumerate(titles, 1):
        print(f"    {i}. {t}")

# **kwargs — variable keyword arguments
def create_movie_record(**kwargs):
    """Build a movie dict from keyword arguments."""
    record = {k: v for k, v in kwargs.items()}
    return record

# Combined
def flexible_filter(*args, min_rating=6.0, **metadata):
    """Filter movies above min_rating, attach metadata."""
    passed = [r for r in args if r >= min_rating]
    return {'passed': passed, 'metadata': metadata, 'count': len(passed)}

print("── *args ──")
movie_summary('Inception', 'Avatar', 'The Dark Knight', 'Titanic')

print("\n── **kwargs ──")
rec = create_movie_record(title='Interstellar', year=2014, rating=8.6, genre='Sci-Fi')
print(f"  Record: {rec}")

print("\n── *args + **kwargs combined ──")
result = flexible_filter(8.8, 4.5, 7.2, 3.1, 9.0, min_rating=7.0, source='TMDB', year=2023)
print(f"  Result: {result}")
```

```python

#  3.3  Recursive Function


def count_high_rated(ratings_list, threshold=7.0, idx=0, count=0):
    """Recursively count movies above threshold."""
    if idx == len(ratings_list):
        return count   # base case
    new_count = count + (1 if ratings_list[idx] >= threshold else 0)
    return count_high_rated(ratings_list, threshold, idx + 1, new_count)  # recursive case

sample = [8.8, 3.2, 6.5, 9.1, 4.8, 7.4, 5.0, 7.8, 6.0, 8.2]
result = count_high_rated(sample, threshold=7.0)
print(f"Recursive count of movies ≥ 7.0 in {sample}: {result}")
```

```python

#  3.4  Lambda · Map · Filter · Reduce · Zip · Enumerate


movies = ['inception', 'avatar', 'the dark knight', 'titanic', 'interstellar']
ratings = [8.8, 7.9, 9.0, 7.8, 8.6]
years   = [2010, 2009, 2008, 1997, 2014]

# Lambda
to_title     = lambda s: s.title()
scale_rating = lambda r: round(r / 10 * 100, 1)

print("── Lambda ──")
print(f"  to_title('inception')  : {to_title('inception')}")
print(f"  scale_rating(8.8)      : {scale_rating(8.8)}%")

# Map
titled = list(map(to_title, movies))
scaled = list(map(scale_rating, ratings))
print("\n── Map ──")
print(f"  titles  : {titled}")
print(f"  scaled  : {scaled}")

# Filter
high   = list(filter(lambda r: r >= 8.5, ratings))
modern = list(filter(lambda y: y >= 2009, years))
print("\n── Filter ──")
print(f"  ratings ≥ 8.5 : {high}")
print(f"  years ≥ 2009  : {modern}")

# Reduce
total_rating = reduce(lambda acc, r: acc + r, ratings)
avg_rating_  = total_rating / len(ratings)
print("\n── Reduce ──")
print(f"  Sum of ratings : {total_rating:.1f}")
print(f"  Average rating : {avg_rating_:.2f}")

# Zip
combined = list(zip(titled, ratings, years))
print("\n── Zip ──")
for name, r, y in combined:
    print(f"  {name:<20} {r:.1f}  ({y})")

# Enumerate
print("\n── Enumerate ──")
for rank, (name, r, y) in enumerate(sorted(combined, key=lambda x: x[1], reverse=True), 1):
    print(f"  #{rank}  {name:<20}  {r}  ({y})")
```

```python

#  5.5  Data Cleaning

movies = df.copy()

# ── 1. Drop duplicates ──
before = len(movies)
movies.drop_duplicates(subset=['title'], keep='first', inplace=True)
movies.reset_index(drop=True, inplace=True)
print(f"Duplicates removed   : {before - len(movies):,}")

# ── 2. Select features ──
keep = [c for c in ['id','title','genres','overview','popularity','vote_average',
                     'vote_count','release_date','revenue','runtime','original_language','status'] if c in movies.columns]
movies = movies[keep]

# ── 3. Handle missing values ──
movies['overview']   = movies['overview'].fillna('No overview available')
movies['genres']     = movies['genres'].fillna('Unknown')
movies['popularity'] = movies['popularity'].fillna(movies['popularity'].median())
movies['vote_average'] = movies['vote_average'].fillna(movies['vote_average'].median())
movies['vote_count'] = movies['vote_count'].fillna(0)
if 'runtime' in movies.columns:
    movies['runtime'] = movies['runtime'].fillna(movies['runtime'].median())
if 'revenue' in movies.columns:
    movies['revenue'] = movies['revenue'].fillna(0)
movies.dropna(subset=['title'], inplace=True)

# ── 4. Type conversions ──
movies['release_date'] = pd.to_datetime(movies['release_date'], errors='coerce')
movies['year']  = movies['release_date'].dt.year
movies['month'] = movies['release_date'].dt.month
for col in ['popularity','vote_average','vote_count']:
    movies[col] = pd.to_numeric(movies[col], errors='coerce').fillna(0)

# ── 5. Remove implausible values ──
movies = movies[(movies['vote_average'] >= 0) & (movies['vote_average'] <= 10)]
movies = movies[movies['popularity'] >= 0]
movies.reset_index(drop=True, inplace=True)

print(f"Final cleaned shape  : {movies.shape}")
print(f"Remaining NaNs       : {movies.isnull().sum().sum()}")
movies.head(3)
```

```python
# ═══════════════════════════════════════════════════════════
#  5.6  Data Normalization
# ═══════════════════════════════════════════════════════════

print("── Min-Max Normalization ──")
for col in ['popularity', 'vote_average', 'vote_count']:
    col_min = movies[col].min()
    col_max = movies[col].max()
    movies[f'{col}_norm'] = (movies[col] - col_min) / (col_max - col_min + 1e-9)
    print(f"  {col}_norm : min={movies[f'{col}_norm'].min():.4f}  max={movies[f'{col}_norm'].max():.4f}")

print("\n── Z-Score Standardization ──")
for col in ['popularity', 'vote_average']:
    mu = movies[col].mean()
    sigma = movies[col].std()
    movies[f'{col}_zscore'] = (movies[col] - mu) / (sigma + 1e-9)
    print(f"  {col}_zscore : mean≈{movies[f'{col}_zscore'].mean():.4f}  std≈{movies[f'{col}_zscore'].std():.4f}")
```

```python
# ═══════════════════════════════════════════════════════════
#  5.7  Exploratory Data Analysis & GroupBy
# ═══════════════════════════════════════════════════════════

print("── Top 5 by Popularity ──")
print(movies.nlargest(5, 'popularity')[['title','year','vote_average','popularity']].to_string(index=False))

print("\n── Top 5 by Rating (vote_count ≥ 100) ──")
well_voted = movies[movies['vote_count'] >= 100]
print(well_voted.nlargest(5, 'vote_average')[['title','year','vote_average','vote_count']].to_string(index=False))

# GroupBy — language
print("\n── GroupBy: original_language ──")
if 'original_language' in movies.columns:
    lang_group = (
        movies.groupby('original_language')
              .agg(count=('title','count'),
                   avg_rating=('vote_average','mean'),
                   avg_popularity=('popularity','mean'))
              .sort_values('count', ascending=False)
              .head(10)
              .round(2)
    )
    print(lang_group)

# GroupBy — year (2000+)
print("\n── GroupBy: Year (2000–2024) ──")
year_group = (
    movies[(movies['year'] >= 2000) & (movies['year'] <= 2024)]
    .groupby('year')
    .agg(movies_released=('title','count'),
         avg_rating=('vote_average','mean'),
         total_popularity=('popularity','sum'))
    .sort_values('year')
    .round(2)
)
print(year_group.tail(10))
```

```python
# ═══════════════════════════════════════════════════════════
#  6.3  Bar Chart — Top Genres
# ═══════════════════════════════════════════════════════════

# Extract all individual genres
all_genres = []
for g in movies['genres'].dropna():
    if isinstance(g, str) and g not in ('Unknown', ''):
        if g.strip().startswith('['):
            try:
                items = ast.literal_eval(g)
                if items and isinstance(items[0], dict):
                    all_genres.extend([i.get('name','') for i in items])
                else:
                    all_genres.extend([str(i) for i in items])
            except Exception:
                all_genres.extend([x.strip() for x in g.split(',') if x.strip()])
        else:
            all_genres.extend([x.strip() for x in g.split(',') if x.strip()])

genre_counts = Counter(all_genres)
top_genres   = pd.Series(genre_counts).sort_values(ascending=False).head(15)

fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Vertical bar
colors = sns.color_palette('husl', len(top_genres))
axes[0].bar(top_genres.index, top_genres.values, color=colors, edgecolor='white')
axes[0].set_title('Top 15 Genres — Vertical Bar', fontweight='bold')
axes[0].set_xlabel('Genre'); axes[0].set_ylabel('Count')
axes[0].tick_params(axis='x', rotation=45)

# Horizontal bar
axes[1].barh(top_genres.index[::-1], top_genres.values[::-1],
             color=colors[::-1], edgecolor='white')
axes[1].set_title('Top 15 Genres — Horizontal Bar', fontweight='bold')
axes[1].set_xlabel('Count')
for i, v in enumerate(top_genres.values[::-1]):
    axes[1].text(v + 10, i, f'{v:,}', va='center', fontsize=8)

plt.tight_layout(); save_plot('viz_03_bar'); plt.show()
```

```python

#  6.5  Waffle Chart — Genre Distribution

# Manual waffle chart (no external library needed)
top8 = top_genres.head(8)
total = top8.sum()
proportions = (top8 / total * 100).round().astype(int)

# Build 10×10 grid
waffle_data = []
for genre, count in proportions.items():
    waffle_data.extend([genre] * count)
waffle_data = waffle_data[:100]  # cap at 100 cells
while len(waffle_data) < 100:
    waffle_data.append(waffle_data[-1])

grid = np.array(waffle_data).reshape(10, 10)
unique_genres_w = top8.index.tolist()
palette = sns.color_palette('husl', len(unique_genres_w))
color_map = {g: palette[i] for i, g in enumerate(unique_genres_w)}

fig, ax = plt.subplots(figsize=(10, 7))
for r in range(10):
    for c in range(10):
        genre = grid[r, c]
        color = color_map.get(genre, (0.7, 0.7, 0.7))
        rect = mpatches.FancyBboxPatch(
            (c * 1.1, (9 - r) * 1.1), 0.9, 0.9,
            boxstyle='round,pad=0.05',
            facecolor=color, edgecolor='white', linewidth=1.5
        )
        ax.add_patch(rect)

legend_patches = [mpatches.Patch(color=color_map[g], label=f"{g} ({proportions.get(g,0)}%)") for g in unique_genres_w]
ax.legend(handles=legend_patches, loc='lower center', bbox_to_anchor=(0.5, -0.18),
          ncol=4, fontsize=9, frameon=False)
ax.set_xlim(-0.2, 11.5); ax.set_ylim(-0.2, 11.5)
ax.set_aspect('equal'); ax.axis('off')
ax.set_title('Waffle Chart — Genre Distribution (Top 8 Genres, 1 cell = 1%)',
             fontsize=13, fontweight='bold', pad=15)
plt.tight_layout(); save_plot('viz_05_waffle'); plt.show()
```

```python
# ═══════════════════════════════════════════════════════════
#  6.7  Seaborn — Multiple Plot Types
# ═══════════════════════════════════════════════════════════

plot_df = movies[
    (movies['vote_average'] > 0) &
    (movies['popularity'] > 0) &
    (movies['vote_count'] >= 50)
].sample(min(3000, len(movies[
    (movies['vote_average'] > 0) &
    (movies['popularity'] > 0) &
    (movies['vote_count'] >= 50)
])), random_state=42)

# --- 6.7a Box Plot by decade ---
movies_decade = movies.dropna(subset=['year']).copy()
movies_decade = movies_decade[
    (movies_decade['year'] >= 1970) & (movies_decade['year'] <= 2024) &
    (movies_decade['vote_average'] > 0)
].copy()
movies_decade['decade'] = (movies_decade['year'] // 10 * 10).astype(int).astype(str) + 's'

fig, axes = plt.subplots(2, 2, figsize=(16, 12))

sns.boxplot(data=movies_decade, x='decade', y='vote_average',
            palette='husl', ax=axes[0,0])
axes[0,0].set_title('Rating Distribution by Decade', fontweight='bold')
axes[0,0].set_xlabel('Decade'); axes[0,0].set_ylabel('Vote Average')

# --- 6.7b Violin Plot ---
sns.violinplot(data=movies_decade, x='decade', y='vote_average',
               palette='muted', inner='quartile', ax=axes[0,1])
axes[0,1].set_title('Violin Plot — Rating by Decade', fontweight='bold')
axes[0,1].set_xlabel('Decade'); axes[0,1].set_ylabel('Vote Average')

# --- 6.7c Scatter + Hue ---
sns.scatterplot(
    data=plot_df.sample(500, random_state=1),
    x='vote_average', y='popularity',
    alpha=0.5, s=20, color='steelblue', ax=axes[1,0]
)
axes[1,0].set_title('Popularity vs Rating (Scatter)', fontweight='bold')
axes[1,0].set_xlabel('Vote Average'); axes[1,0].set_ylabel('Popularity')
axes[1,0].set_ylim(0, plot_df['popularity'].quantile(0.97))

# --- 6.7d Heatmap (Correlation) ---
num_cols = ['popularity','vote_average','vote_count']
if 'runtime' in movies.columns: num_cols.append('runtime')
if 'revenue' in movies.columns: num_cols.append('revenue')
if 'year'    in movies.columns: num_cols.append('year')
corr = movies[num_cols].replace(0, np.nan).dropna().corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm',
            center=0, mask=mask, ax=axes[1,1],
            linewidths=0.5, annot_kws={'size': 10})
axes[1,1].set_title('Feature Correlation Heatmap', fontweight='bold')

plt.suptitle('Seaborn Visualizations — TMDB Dataset', fontsize=16, fontweight='bold', y=1.01)
plt.tight_layout(); save_plot('viz_07_seaborn'); plt.show()
```
